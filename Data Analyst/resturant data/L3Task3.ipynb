```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '/content/Dataset .csv'
data = pd.read_csv(file_path)


data.dropna(subset=['Price_range', 'Has Online delivery', 'Has Table booking'], inplace=True)


online_delivery_vs_price = pd.crosstab(data['Price_range'], data['Has Online delivery'], normalize='index') * 100
table_booking_vs_price = pd.crosstab(data['Price_range'], data['Has Table booking'], normalize='index') * 100


plt.figure(figsize=(8, 6))
online_delivery_vs_price.plot(kind='bar', stacked=True, color=['lightblue', 'lightgreen'])
plt.title('Has Online delivery vs Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=45)
plt.legend(title='Has Online delivery', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
table_booking_vs_price.plot(kind='bar', stacked=True, color=['lightcoral', 'lightseagreen'])
plt.title('Has Table booking vs Price_range')
plt.xlabel('Price_range')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=45)
plt.legend(title='Has Table booking', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()

online_delivery_percentage = online_delivery_vs_price['Yes']
table_booking_percentage = table_booking_vs_price['Yes']

print("Percentage of restaurants offering online delivery by price range:")
print(online_delivery_percentage)
print("\nPercentage of restaurants offering table booking by price range:")
print(table_booking_percentage)
```
