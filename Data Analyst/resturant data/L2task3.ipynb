```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# Step 1: Load the dataset
file_path = '/content/Dataset .csv'  # Update with the correct path to your dataset
data = pd.read_csv(file_path)

# Step 2: Data Cleaning (Ensure no missing values in 'longitude' and 'latitude' columns)
data.dropna(subset=['Longitude', 'Latitude'], inplace=True)

# Step 3: Create a map centered around the average location
map_center = [data['Latitude'].mean(), data['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)

# Step 4: Add a marker cluster to visualize restaurant locations
marker_cluster = MarkerCluster().add_to(restaurant_map)

# Step 5: Add markers to the map for each restaurant location
for idx, row in data.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['name'] if 'name' in data.columns else None  # Show restaurant name on hover if available
    ).add_to(marker_cluster)

# Step 6: Save or display the map
restaurant_map.save('restaurant_locations_map.html')  # Save the map as an HTML file
restaurant_map  # Display the map in the notebook
```
