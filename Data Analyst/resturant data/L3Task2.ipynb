```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/Dataset .csv'
data = pd.read_csv(file_path)


data.dropna(subset=['Votes', 'Aggregate rating'], inplace=True)

max_Votes = data.loc[data['Votes'].idxmax()]
min_Votes = data.loc[data['Votes'].idxmin()]

print("Restaurant with the highest number of Votes:")
print(max_Votes[['Restaurant Name', 'Votes', 'Aggregate rating']])

print("\nRestaurant with the lowest number of Votes:")
print(min_Votes[['Restaurant Name', 'Votes', 'Aggregate rating']])

correlation = data['Votes'].corr(data['Aggregate rating'])
print(f"\nCorrelation between number of Votes and Aggregate rating: {correlation:.2f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=data, color='blue')
plt.title('Correlation Between Number of Votes and Aggregate rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate rating')
plt.tight_layout()
plt.show()

sns.lmplot(x='Votes', y='Aggregate rating', data=data, height=6, aspect=1.5, line_kws={"color": "red"})
plt.title('Votes vs Aggregate rating with Regression Line')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate rating')
plt.show()
```
