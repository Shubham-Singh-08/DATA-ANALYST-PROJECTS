```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '/content/Dataset .csv'
data = pd.read_csv(file_path)


data.dropna(subset=['Price_range'], inplace=True)


price_range_counts = data['Price_range'].value_counts()


total_restaurants = len(data)-1
Price_range_percentages = (price_range_counts / total_restaurants) * 100


plt.figure(figsize=(8, 6))
price_range_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Restaurants by Price_range')
plt.xlabel('Number of Restaurants')
plt.ylabel('Price_range')
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()


print("Percentage of restaurants in each price range:")
for Price_range, percent in Price_range_percentages.items():
    print(f"{Price_range}: {percent:.2f}%")


```
